```python
def check_valid(row,col):
    entry = row[col]
    if entry >= 0:
        return entry
    else: return np.nan
```

```python
for col in df.columns[1:]:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    # axis=1 column-wise operation
    df[col] = df.apply(check_valid,axis=1,col=col)
df.head()
```

## Reshaping Data

Currently, the gapminder DataFrame has a separate column for each year. What you want instead is a single column that contains the year, and a single column that represents the average life expectancy for each year and country. By having year in its own column, you can use it as a predictor variable in a later analysis.

```python
df_melt = pd.melt(df,id_vars='Life expectancy')

# Rename the columns
df_melt.columns = ['country','year','life_expectancy']

df_melt.head()
```

## Looking at country spellings

```python
df_melt.country
```

```python
df_melt.country.value_counts()
```

It is reasonable to assume that country names will contain:

* The set of lower and upper case letters.
* Whitespace between words.
* Periods for any abbreviations.

To confirm that the column is fine we need to use regular expressions

```python
countries = df_melt['country']

# Drop all the duplicates from countries
countries = countries.drop_duplicates()

# Write the regular expression: pattern
pattern = '^[A-Za-z\.\s]*$'

# Create the Boolean vector: mask
mask = countries.str.contains(pattern)

# Invert the mask: mask_inverse
# to have a Boolean for what didn't match
mask_inverse = ~mask
print(mask_inverse.head())

# Subset countries using mask_inverse: invalid_countries
invalid_countries = countries.loc[mask_inverse]

# Print invalid_countries
invalid_countries
```

```python
df_clean = df_melt.drop(invalid_countries.index, axis=0)

print(df_melt.shape)
print(df_clean.shape)
```
