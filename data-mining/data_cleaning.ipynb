## Reshaping Data

Its the way we organize the data into a formal and standard form. The principles of tidy data are:
* Columns represent separate variables
* Rows represent indivisual observations
* Observational units form tables

### Reshaping your data using melting
Melting data is the process of turning columns of your data into rows of data. 

This exercise, you will practice melting a DataFrame using:
* **`pd.melt()`**.
    * **`id_vars`**: which col to be index (you dont want to melt them)
    * **`value_vars`** cols you want to melt into rows
    
By default, if no value_vars are provided, all columns not set in the `id_vars` will be melted.

In the example below, the airquality dataframe is already tidy! but we will show how `df.melt` can untidy it (and of course vise versa!). In other words, we will melt its `Ozone`, `Solar.R`, `Wind`, and `Temp` columns into rows. 

````python
airquality = pd.read_csv('airquality.csv')
airquality.head()
````

````python
# Melt airquality: airquality_melt
airquality_melt = pd.melt(airquality, id_vars=['Month', 'Day'])

# Print the head of airquality_melt
airquality_melt.head()
````

### Customizing melted data

When melting DataFrames, it would be better to have column names more meaningful than `variable` and `value` (the default names used by `pd.melt()`).

* **`var_name`**
* **`value_name`**

````python
# Melt airquality: airquality_melt
airquality_melt = pd.melt(airquality, id_vars=['Month', 'Day'], var_name='measurement', value_name='reading')

# Print the head of airquality_melt
airquality_melt.head()
````

## Pivoting Data

* Opposite to melting
* Turn unique values into separate columns
* Analysis freindly, you can play with the shape of the data for further analysis
* Data becomes tidy

* **`.pivot_table()`** / **`.pivot()`**
    * **`index`** parameter which you can use to specify the columns that you don't want pivoted: It is similar to the `id_vars` parameter of `pd.melt()`. 
    * **`columns`** the name of the column you want to pivot
    * **`values`** the values to be used when the column is pivoted.
    
The difference between `pivot` and `pivot_table` is that `pivot` requires unique columns as indexes

````python
# print data before pivoting
print(airquality_melt.head())

# Pivot airquality_melt: airquality_pivot
airquality_pivot = airquality_melt.pivot_table(index=['Month','Day'], columns='measurement', values='reading')

# Print the head of airquality_pivot
airquality_pivot.head()
````

Notice the mulitevel index here (`Month` and `Day`) you can reset it using:
* **`.reset_index()`**

````python
# Print the index of airquality_pivot
print(airquality_pivot.index)

# Reset the index of airquality_pivot: airquality_pivot_reset
airquality_pivot_reset = airquality_pivot.reset_index()

# Print the new index of airquality_pivot_reset
print(airquality_pivot_reset.index)

# Print the head of airquality_pivot_reset
airquality_pivot_reset.head()
````

### Pivoting duplicate values

* **`aggfunc`** parameter in **`.pivot_table()`**

````python
airquality_pivot = airquality_melt.pivot_table(index=['Month', 'Day'], columns='measurement', values='reading', aggfunc=np.mean)
airquality_pivot.head()
````

**NOTE** The default aggregation function used in `.pivot_table()` is `np.mean()`. So you could have pivoted the duplicate values in this DataFrame even without explicitly specifying the `aggfunc` parameter. Note that in this example there is no duplicate observations; each pair of index has a unique observation. Thus, using aggregation functions will result on the same values.

Overall, the `pivot_table` is a pefect reshaping tool for relating all variables with one or two indexes.

Other aggregation functions:
* **`aggfunc='count'`**
* **`aggfunc=sum`**

````python
airquality_pivot = airquality_melt.pivot_table(index=['Month', 'Day'], columns='measurement', values='reading', aggfunc='count')
airquality_pivot.head()
````

### Using margins in pivot tables

* **`margins=True`**: to add totals in the margins of pivot table

````python
airquality_pivot = airquality_melt.pivot_table(index=['Month', 'Day'], columns='measurement', values='reading', aggfunc=sum ,margins=True)
airquality_pivot.head()
````

### Stacking and Unstacking level indexes

* **`.stack()`**: Stack the prescribed level(s) from columns to index
    * **`level`**
    * **`dropna`**
* **`.unstack()`**: Pivot a level of the (necessarily hierarchical) index labels, returning a DataFrame having a new level of column labels whose inner-most level consists of the pivoted index labels.
    * **`level`**
    * **`fill_value`**
    
    
**NOTE:** **`.unstack()`** can be used to transform a multi-level indexed series (might be a result of a certian groupby operation) into a dataframe 

````python
users = pd.read_csv('users.csv', index_col=['city', 'weekday'])
del users['Unnamed: 0']

users.head()
````

Notice that the dataframe has 2 levels of index, level `0` is `city` and level `1` is `weekday`

````python
# Unstack users by 'weekday': byweekday
byweekday = users.unstack(level='weekday')

byweekday
````

If you `stack` the df it will go back to what it was

````python
byweekday.stack(level='weekday')
````

Let's unstack and stack the dataset by `city` which is the outer index. Notice when it stacked back into the data it became 2nd index.

````python
# Unstack users by 'city': bycity
bycity = users.unstack(level='city')

# Print the bycity DataFrame
print(bycity)

# Stack bycity by 'city' and print it
newusers = bycity.stack(level='city')
newusers
````

You can use **`swaplevel(0,1)`** to get the index order back

````python
newusers.swaplevel(0,1)
````

## Splitting Datas
### Splitting a column with `.str`

````python
import pandas as pd

tb = pd.read_csv('tb.csv')
tb.head()
````

One of the problems in this dataset is that for example `m014` represents males aged 0-14 years of age. In order to parse this value, you need to extract the first letter into a new column for `gender`, and the rest into a column for `age_group`. This will be done by slicing using `.str` attribute of columns of type `object`

````python
# Melt tb: tb_melt
tb_melt = pd.melt(tb, id_vars=['country', 'year'])

# Create the 'gender' column
tb_melt['gender'] = tb_melt.variable.str[0] #column 'variable'

# Create the 'age_group' column
tb_melt['age_group'] = tb_melt.variable.str[1:]

# Print the head of tb_melt
tb_melt.head()
````

### Splitting a column with `.split()` and `.get()`

````python
import pandas as pd 

ebola = pd.read_csv('ebola.csv')
ebola.head()
````

````python
ebola.columns
````

Notice that the data has column names such as `Cases_Guinea` and `Deaths_Guinea`. Here, the underscore `_` serves as a delimiter between the first part (cases or deaths), and the second part (country).

````python
# Melt ebola: ebola_melt
ebola_melt = pd.melt(ebola, id_vars=['Date', 'Day'], var_name='type_country', value_name='counts')

ebola_melt.head()
````

Now we need to clean `date` and `time` columns

* `date` col is converted to string using **`.astype(str)`**
* `time` col is formated as a 4 digit format (zeroes will be padded to left) using **`{:0>4}.format`** in a lambda function. To know more about `format` function you can refer to its [documentation](https://pyformat.info/)
* `date` and `time` columns are concatenated to each other
* the result string is converted into datetime using **`.to_datetime()`** with the format: `'%Y%m%d%H%M'`
* this column is set as an index using **`.set_index`**

````python
# Convert the date column to string: df['date']
df['date'] = df['date'].astype(str)
df['date'].head(2)
````

````python
# Pad leading zeros to the Time column: df['Time']
df['Time'] = df['Time'].apply(lambda x:'{:0>4}'.format(x))
df['Time'].head(2)
````

````python
# Concatenate the new date and Time columns: date_string
date_string = df['date'] + df['Time']

date_string.head(2)
````

````python
# Convert the date_string Series to datetime: date_times
date_times = pd.to_datetime(date_string, format='%Y%m%d%H%M')

# Set the index to be the new date_times container: df_clean
df_clean = df.set_index(date_times)

# Print the output of df_clean.head()
df_clean.head()
````
