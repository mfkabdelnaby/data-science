## Data Visualization for further diagnosis

### Histograms

`df.plot`, and the `kind` parameter allows you to specify the type of plot to use. For **Histograms** use - `kind='hist'`

```python
# Import matplotlib.pyplot
import matplotlib.pyplot as plt

# Describe the column
df['Existing Zoning Sqft'].describe()
```

```python
# Plot the histogram
df['Existing Zoning Sqft'].plot(kind='hist')

# Display the histogram
plt.show()
```

You'll notice that there are extremely large differences between the `min` and `max` values, and the plot will need to be adjusted accordingly. In such cases, it's good to look at the plot on a *log* scale. The keyword arguments `logx=True` or `logy=True` can be passed in to `.plot()` depending on which axis you want to rescale.

```python
# Plot the histogram
df['Existing Zoning Sqft'].plot(kind='hist', logx=True, logy=True)

# Display the histogram
plt.show()
```

As you saw here, you still needed to look at the summary statistics to help understand your data better. You expected a large amount of counts on the left side of the plot because the 25th, 50th, and 75th percentiles have a value of `0`. The plot shows us that there are barely any counts near the `max` value, signifying an outlier.

### boxplots

Histograms are great ways of visualizing single variables. To visualize multiple variables, **boxplots** are useful, especially when one of the variables is categorical. In other words, boxplots are great when you have a numeric column that you want to compare across different categories. 

`.boxplot()` in which you have to specify the `column` and `by` parameters. 

```python
# Import necessary modules
import pandas as pd
import matplotlib.pyplot as plt

# Create the boxplot
df.boxplot(column='Proposed Zoning Sqft', by='Borough', rot=90)
# rotate the label by 90 degrees

# Display the plot
plt.show()
```

You can see the 2 extreme outliers are in the borough of Manhattan. An initial guess could be that since land in Manhattan is extremely expensive, these outliers may be valid data points. Again, further investigation is needed to determine whether or not you can drop or keep those points in your data.

## Tidy Data

Its the way we organize the data into a formal and standard form. The principles of tidy data are:
* Columns represent separate variables
* Rows represent indivisual observations
* Observational units form tables

### Reshaping your data using melting
Melting data is the process of turning columns of your data into rows of data. 

This exercise, you will practice melting a DataFrame using `pd.melt()`. There are two parameters you should be aware of: `id_vars` and `value_vars`. The `id_vars` represent the columns of the data you do not want to melt (i.e., keep it in its current shape), while the `value_vars` represent the columns you do wish to melt into rows. By default, if no value_vars are provided, all columns not set in the `id_vars` will be melted.

In the example below, the airquality dataframe is already tidy! but we will show how `df.melt` can untidy it (and of course vise versa!). In other words, we will melt its `Ozone`, `Solar.R`, `Wind`, and `Temp` columns into rows. 

```python
airquality = pd.read_csv('airquality.csv')
airquality.head()
```

```python
# Melt airquality: airquality_melt
airquality_melt = pd.melt(airquality, id_vars=['Month', 'Day'])

# Print the head of airquality_melt
airquality_melt
```

### Customizing melted data

When melting DataFrames, it would be better to have column names more meaningful than `variable` and `value` (the default names used by `pd.melt()`).

* `var_name`
* `value_name`

```python
# Melt airquality: airquality_melt
airquality_melt = pd.melt(airquality, id_vars=['Month', 'Day'], var_name='measurement', value_name='reading')

# Print the head of airquality_melt
airquality_melt.head()
```

### Pivoting Data

* Opposite to melting
* Turn unique values into separate columns
* From analysis friendly shape to reporting friendly shape
* Data becomes tidy

`.pivot_table()` has an `index` parameter which you can use to specify the columns that you don't want pivoted: It is similar to the `id_vars` parameter of `pd.melt()`. Two other parameters that you have to specify are `columns` (the name of the column you want to pivot), and `values` (the values to be used when the column is pivoted).

```python
# Print the head of airquality_melt
airquality_melt.head()

# Pivot airquality_melt: airquality_pivot
airquality_pivot = airquality_melt.pivot_table(index=['Month', 'Day'], columns='measurement', values='reading')

# Print the head of airquality_pivot
airquality_pivot.head()
```

Notice that the dataframe doesnt look quite like the original one. What you got back instead was a pandas DataFrame with a hierarchical index (also known as a MultiIndex). This topic will be explained in detail later.

To get back the original dataframe we use:
* `.reset_index()`

```python
# Print the index of airquality_pivot
print(airquality_pivot.index)

# Reset the index of airquality_pivot: airquality_pivot_reset
airquality_pivot_reset = airquality_pivot.reset_index()

# Print the new index of airquality_pivot_reset
print(airquality_pivot_reset.index)

# Print the head of airquality_pivot_reset
airquality_pivot_reset.head()
```

### Pivoting duplicate values

* `aggfunc` parameter in `.pivot_table()`

Note that the code below will not work since the dataframe `airquality` is already pivotted, plus it has no duplicates.


```python
import numpy as np

airquality_pivot = airquality.pivot_table(index=['Month', 'Day'], columns='measurement', values='reading', aggfunc=np.mean)

# Print the head of airquality_pivot before reset_index
print(airquality_pivot.head())

# Reset the index of airquality_pivot
airquality_pivot = airquality_pivot.reset_index()

# Print the head of airquality_pivot
print(airquality_pivot.head())

# Print the head of airquality
print(airquality.head())
```

**NOTE** The default aggregation function used in `.pivot_table()` is `np.mean()`. So you could have pivoted the duplicate values in this DataFrame even without explicitly specifying the `aggfunc` parameter.

### Splitting a column with .str

```python
import pandas as pd

tb = pd.read_csv('tb.csv')
tb.head()
```

One of the problems in this dataset is that for example `m014` represents males aged 0-14 years of age. In order to parse this value, you need to extract the first letter into a new column for `gender`, and the rest into a column for `age_group`. This will be done by slicing using `.str` attribute of columns of type `object`

```python
# Melt tb: tb_melt
tb_melt = pd.melt(tb, id_vars=['country', 'year'])

# Create the 'gender' column
tb_melt['gender'] = tb_melt.variable.str[0] #column 'variable'

# Create the 'age_group' column
tb_melt['age_group'] = tb_melt.variable.str[1:]

# Print the head of tb_melt
tb_melt.head()
```
